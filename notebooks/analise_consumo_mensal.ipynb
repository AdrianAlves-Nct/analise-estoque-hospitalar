## Visualização Inicial DF Consumo Mensal

```python
import pandas as pd
```

```python
df_consumo_mensal = pd.read_csv("../dados/consumo_mensal.csv", sep=",")
display(df_consumo_mensal.head())
display(df_consumo_mensal.tail())
```

```python
df_consumo_mensal.info()
df_consumo_mensal.describe()
```

## Tratamento da Coluna Data do DF Consumo Mensal

- Objetivo: transformar a coluna para o formato datetime64 e criar colunas auxiliares como mês.

```python
df_consumo_mensal["Data"] = pd.to_datetime(df_consumo_mensal["Data"], format="%Y-%m-%d")
df_consumo_mensal["Mes"] = df_consumo_mensal["Data"].dt.month_name()
display(df_consumo_mensal.head())
df_consumo_mensal.info()
```

## Verificação de Nulos e Duplicados DF Consumo Mensal

```python
display(df_consumo_mensal.isna().sum())
display(df_consumo_mensal.duplicated().sum())
```

## Validação da Quantidade Consumida DF Consumo Mensal

- Regra básica:

Quantidade consumida deve ser maior que zero

```python
condicao_consumo = df_consumo_mensal["Quantidade_Consumida"] < 0
display(df_consumo_mensal[condicao_consumo])
```

## Padronização de Setor DF Consumo Mensal

- Verificar se há valores únicos de Setor

```python
setores_unicos = df_consumo_mensal["Setor"].unique()
display(setores_unicos)
```

## Análise Temporal Inicial DF Consumo Mensal

- Consumo total por mês

```python
dicionario_meses_pt = {
    "January": "Janeiro",
    "February": "Fevereiro",
    "March": "Março",
    "April": "Abril",
    "May": "Maio",
    "June": "Junho",
    "July": "Julho",
    "August": "Agosto",
    "September": "Setembro",
    "October": "Outubro",
    "November": "Novembro",
    "December": "Dezembro"
}
df_consumo_mensal["Mes"] = df_consumo_mensal["Mes"].map(dicionario_meses_pt)
display(df_consumo_mensal.head())
```

```python
lista_meses_pt = [
    "Janeiro",
    "Fevereiro",
    "Março",
    "Abril",
    "Maio",
    "Junho",
    "Julho",
    "Agosto",
    "Setembro",
    "Outubro",
    "Novembro",
    "Dezembro"
]

df_consumo_total_mes = df_consumo_mensal.groupby(["Mes"]).agg(
    Quantidade_Total=("Quantidade_Consumida", "sum")
).reset_index()

df_consumo_total_mes["Mes"] = pd.Categorical(df_consumo_total_mes["Mes"], categories=lista_meses_pt, ordered=True)

df_consumo_total_mes = df_consumo_total_mes.sort_values("Mes").reset_index(drop=True)

display(df_consumo_total_mes)
```

- Consumo total por setor

```python

df_consumo_total_setor = df_consumo_mensal.groupby(["Setor"]).agg(
    Quantidade_Total=("Quantidade_Consumida", "sum")
).sort_values(by="Quantidade_Total", ascending=False).reset_index()
display(df_consumo_total_setor)
```

- Consumo Mensal por Setor

```python
df_consumo_mensal_setor = df_consumo_mensal.groupby(["Mes", "Setor"]).agg(
    Quantidade_Total=("Quantidade_Consumida", "sum")
).reset_index()

df_consumo_mensal_setor["Mes"] = pd.Categorical(df_consumo_mensal_setor["Mes"], categories=lista_meses_pt, ordered=True)

df_consumo_mensal_setor = df_consumo_mensal_setor.sort_values(by=["Mes", "Quantidade_Total"], ascending=[True, False]).reset_index(drop=True)

display(df_consumo_mensal_setor)
```

- Consumo Total por Grupo

```python
df_consumo_total_grupo = df_consumo_mensal.groupby("Grupo").agg(
    Quantidade_Total=("Quantidade_Consumida", "sum")
).sort_values(by="Quantidade_Total", ascending=False).reset_index()
display(df_consumo_total_grupo)
```

- Consumo Mensal por Grupo

```python
df_consumo_mensal_grupo = df_consumo_mensal.groupby(["Mes", "Grupo"]).agg(
    Quantidade_Total=("Quantidade_Consumida","sum")
).reset_index()

df_consumo_mensal_grupo["Mes"] = pd.Categorical(df_consumo_mensal_grupo["Mes"], categories=lista_meses_pt, ordered=True)

df_consumo_mensal_grupo = df_consumo_mensal_grupo.sort_values(by=["Mes", "Quantidade_Total"], ascending=[True, False]).reset_index(drop=True)

display(df_consumo_mensal_grupo)
```

- Posso "desmembrar" para ver só um grupo ou só um mês... 

```python
df_consumo_mensal_grupo_medicamentos = df_consumo_mensal_grupo[df_consumo_mensal_grupo["Grupo"] == "Medicamento"]
display(df_consumo_mensal_grupo_medicamentos)
```

- Produtos mais consumidos


```python
df_consumo_total_produto = df_consumo_mensal.groupby("Produto").agg(
    Quantidade_Total_Consumida=("Quantidade_Consumida", "sum")
).sort_values(by="Quantidade_Total_Consumida", ascending=False).reset_index()
display(df_consumo_total_produto)
```

- Produto mais consumido por mês

```python
df_consumo_mensal_produto = df_consumo_mensal.groupby(["Mes", "Produto"]).agg(
    Quantidade_Total=("Quantidade_Consumida", "sum")
).reset_index()

df_consumo_mensal_produto["Mes"] = pd.Categorical(df_consumo_mensal_produto["Mes"], categories=lista_meses_pt, ordered=True)

df_consumo_mensal_produto = df_consumo_mensal_produto.sort_values(by=["Mes", "Quantidade_Total"], ascending=[True, False]).reset_index(drop=True)

display(df_consumo_mensal_produto.head(25))
```
